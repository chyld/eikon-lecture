```python
# Reuters Instrument Code
rics = [
    'GE',  # General Electric stock
    'AAPL.O',  # Apple stock
    '.SPX',  # S&P 500 stock index
    '.VIX',  # VIX volatility index
    'EUR=',  # EUR/USD exchange rate
    'XAU=',  # Gold price
]
```

```python
# RIC to International Securities Identification Numbers
ek.get_symbology(rics, from_symbol_type='RIC', to_symbol_type='ISIN')
```

```python
# multiple conversions
dfsym = ek.get_symbology(rics, from_symbol_type='RIC', to_symbol_type=['ISIN', 'ticker'])
dfsym
```

```python
# get apple ric to ticker
dfsym.loc['AAPL.O']['ticker']
```
